# Task
Generate synthetic data for a multiple linear regression with at least two independent variables, perform multiple linear regression using `statsmodels`, and visualize the data and the fitted regression plane in a 3D scatter plot.

## Generate data

### Subtask:
Create synthetic data for a multiple linear regression with at least two independent variables.


**Reasoning**:
The subtask is to create synthetic data for a multiple linear regression. The instructions provide the steps to achieve this, including setting a random seed, generating independent and dependent variables, and storing them in a pandas DataFrame. I will combine these steps into a single code block.



```python
np.random.seed(42)
x_1 = np.random.rand(100)
x_2 = np.random.rand(100)
Y = 2 * x_1 + 3 * x_2 + 5 + np.random.randn(100)
data = pd.DataFrame({'x_1': x_1, 'x_2': x_2, 'Y': Y})
data.head()
```

## Perform multiple linear regression

### Subtask:
Use `statsmodels` to fit a multiple linear regression model to the generated data.


**Reasoning**:
Add a constant to the independent variables and fit the OLS model using statsmodels, then print the model summary.



```python
x_ols = sm.add_constant(data[['x_1', 'x_2']])
model = sm.OLS(data['Y'], x_ols).fit()
model.summary()
```

## Visualize the model

### Subtask:
Generate a 3D scatter plot of the data and the fitted regression plane.


**Reasoning**:
The previous attempt to create a 3D plot failed due to an incorrect keyword argument for `gca()`. This time, I will ensure the correct approach is used to create a 3D axes object and proceed with generating the meshgrid, calculating predicted values, plotting the surface, and setting the labels as per the instructions.



```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(data['x_1'], data['x_2'], data['Y'], c='g', marker='o')

x_1_surf, x_2_surf = np.meshgrid(np.linspace(data['x_1'].min(), data['x_1'].max(), 100),
                                 np.linspace(data['x_2'].min(), data['x_2'].max(), 100))

# Create a DataFrame for prediction using the meshgrid values
surf_data = pd.DataFrame({'x_1': x_1_surf.ravel(), 'x_2': x_2_surf.ravel()})
surf_data_ols = sm.add_constant(surf_data)
y_hat_surf = model.predict(surf_data_ols).values.reshape(x_1_surf.shape)

ax.plot_surface(x_1_surf, x_2_surf, y_hat_surf, color='black', alpha=0.2, linewidth=1, antialiased=True)

ax.set_xlabel('x_1')
ax.set_ylabel('x_2')
ax.set_zlabel('Y, $\hat{Y}$')

plt.show()
```
